Hypothesis 1: Technology products have the highest profit margin compared to other product categories.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Grouping data by category and summing profits
cat_profit = data.groupby('Category')['Profit'].sum()

# Sorting categories by profit in descending order
cat_profit_sorted = cat_profit.sort_values(ascending=False)

# Creating a bar plot
plt.figure(figsize=(10, 6))  # Adjusting figure size for better presentation
cat_profit_sorted.plot(kind='bar', color='skyblue')

# Adding title and labels
plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")

# Adding value labels on top of each bar
for i, profit in enumerate(cat_profit_sorted):
    plt.text(i, profit + 1000, f"${profit:,.0f}", ha='center', va='bottom')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.tight_layout()  # Adjusting layout for better spacing
plt.savefig('profit_by_category1.png')
plt.show()
```

Hypothesis 2: The East region has the highest sales compared to other regions.

```python
# Grouping data by region and summing sales
reg_sales = data.groupby('Region')['Sales'].sum()

# Creating a bar plot
plt.figure(figsize=(10, 6))  # Adjusting figure size for better presentation
reg_sales.plot(kind="bar", color='skyblue')

# Adding title and labels
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")

# Adding value labels on top of each bar
for i, sales in enumerate(reg_sales):
    plt.text(i, sales + 1000, f"${sales:,.0f}", ha='center', va='bottom')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.tight_layout()  # Adjusting layout for better spacing

# Saving the plot as an image file (e.g., PNG)
plt.savefig('sales_by_region.png', dpi=300)  # Set dpi (dots per inch) for high resolution

# Show the plot after saving
plt.show()
```

Hypothesis 3: Sales are higher during certain months of the year.

```python
data.columns
```

```python
# Extracting month from the order date
data['Order Month'] = pd.DatetimeIndex(data['Order Date']).month

# Grouping data by month and summing sales
month_sales = data.groupby('Order Month')['Sales'].sum()

# Creating a line plot
plt.figure(figsize=(10, 6))  # Adjusting figure size for better presentation
month_sales.plot(kind='line', marker='o', color='blue')

# Adding title and labels
plt.title("Total Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")

# Displaying the plot
plt.tight_layout()  # Adjusting layout for better spacing

# Saving the plot as an image file (e.g., PNG)
plt.savefig('sales_by_month3.png')

# Show the plot after saving
plt.show()
```

Hypothesis 4: Orders with same-day shipping have the lowest rate of returned products.

```python
# Grouping data by shipping mode and getting the total number of orders
total_orders_by_shipping_mode = data.groupby('Ship Mode').size()

# Filtering data for returned orders (where profit < 0) and grouping by shipping mode
returned_orders_by_shipping_mode = data[data['Profit'] < 0].groupby('Ship Mode').size()

# Calculating return percentage per shipping mode
returned_per_by_shipping_mode = (returned_orders_by_shipping_mode / total_orders_by_shipping_mode) * 100

# Creating a donut pie chart
plt.figure(figsize=(8, 8))  # Adjusting figure size for better presentation

# Defining a custom color palette
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Plotting the donut pie chart with custom colors
wedges, texts, autotexts = plt.pie(returned_per_by_shipping_mode, labels=returned_per_by_shipping_mode.index,
                                    autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.3))

# Adding a circle at the center to turn the pie chart into a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Adding a title
plt.title("Return Percentage by Shipping Mode")

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show the plot
plt.tight_layout()
# Saving the plot as an image file (e.g., PNG)
plt.savefig('ReturnPercByShippingMode.png', dpi=300)  # Set dpi (dots per inch) for high resolution
plt.show()
```

Hypothesis 5: The company's profit is more on weekdays than on weekends.

```python
# Assuming 'data' is your DataFrame containing order date and profit information

# Extracting the day of the week from the order date
data['Order Day'] = pd.DatetimeIndex(data['Order Date']).day_name()

# Grouping data by day of the week and summing profits
day_profit = data.groupby('Order Day')['Profit'].sum()

# Defining the order of days of the week
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Creating a horizontal bar plot
plt.figure(figsize=(10, 6))  # Adjusting figure size for better presentation
bars = day_profit.loc[order].plot(kind="barh", color='skyblue')

# Adding title and labels
plt.title("Total Profit by Day of the Week")
plt.xlabel("Total Profit")
plt.ylabel("Day of the Week")

# Adding profit values as text annotations on top of each bar
for i, bar in enumerate(bars.patches):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f"${day_profit.iloc[i]:,.2f}",
             va='center', ha='left', color='black')

# Displaying the plot
plt.tight_layout()  # Adjusting layout for better spacing
plt.savefig('Total Profit by Day of the Week5.png')
plt.show()
```
